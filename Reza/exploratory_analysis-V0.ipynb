#### Task 1: 
##### Test the hypothesis that the delay is from Normal distribution. and that mean of the delay is 0. Be careful about the outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(1,2, figsize=(18,4));
sns.distplot(df['arr_delay'], color='b', ax=ax[0]);
ax[0].set_title('Distribution of Arrival Delay Time (Min)', fontsize=14);
ax[0].set_xlim([min(df['arr_delay'].values), max(df['arr_delay'].values)]);

sns.boxplot(df['arr_delay'], ax=ax[1]);
ax[1].set_title('Arrival Delay Time (Min)');
plt.show();
```

```python
df['arr_delay'].describe()
```

```python
def remove_outliers(data, col):
    mean, cutoff = np.mean(data[col]), np.std(data[col]) * 3   
    lower, upper = mean - cutoff, mean + cutoff
    outliers = [x for x in data[col] if x < lower or x > upper]
    df.drop(df[(data[col] > upper) | (data[col] < lower)].index, inplace=True)
    return f'{len(outliers)} outliers removed'
```

```python
remove_outliers(df, 'arr_delay')
```

```python
fig, ax = plt.subplots(1,2, figsize=(18,4));
sns.distplot(df['arr_delay'], color='b', ax=ax[0]);
ax[0].set_title('Distribution of Arrival Delay Time (Min)', fontsize=14);
ax[0].set_xlim([min(df['arr_delay'].values), max(df['arr_delay'].values)]);

sns.boxplot(df['arr_delay'], ax=ax[1]);
ax[1].set_title('Arrival Delay Time (Min)');
plt.show();
```

##### Test if the data is normally distributed

```python
import scipy.stats as st
```

```python
pvalue = st.shapiro(df['arr_delay'])[1]
if pvalue > 0.05:
    print('Data is normal')
elif pvalue < 0.05:
    print('Data deviate from a normal distribution')
```

#### Task 2: 
##### Is average/median monthly delay different during the year? If yes, which are months with the biggest delays and what could be the reason?

```python
from datetime import datetime, date, time
```

```python
df['fl_date'] = pd.to_datetime(df['fl_date'])
df['fl_month'] = df['fl_date'].dt.month

monthly = df[['fl_month','arr_delay']].groupby('fl_month').agg('mean')
monthly.reset_index(inplace=True)
monthly.sort_values('arr_delay', ascending=False)
```

```python
# highest delays occur during the summer time --  Possible reasons : higher traffic / vacations
```

```python
monthly = df[['fl_month','arr_delay','mkt_carrier_fl_num']].groupby('fl_month').agg({'arr_delay':'mean','mkt_carrier_fl_num':'count'})
monthly.reset_index(inplace=True)
monthly.rename(columns={'arr_delay':'avg_arr_delay','mkt_carrier_fl_num':'count_flights'}, inplace=True)
monthly.sort_values('avg_arr_delay', ascending=False)
```

```python
#As shown above during summer we have more flights than other seasons in the year.
```
