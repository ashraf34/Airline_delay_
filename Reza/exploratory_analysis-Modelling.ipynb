```python
def remove_outliers(data, col):
    mean, cutoff = np.mean(data[col]), np.std(data[col]) * 3   
    lower, upper = mean - cutoff, mean + cutoff
    outliers = [x for x in data[col] if x < lower or x > upper]
    df.drop(df[(data[col] > upper) | (data[col] < lower)].index, inplace=True)
    return f'{len(outliers)} outliers removed'
```

```python
remove_outliers(df, 'arr_delay')
```

```python
fig, ax = plt.subplots(1,2, figsize=(18,4));
sns.distplot(df['arr_delay'], color='b', ax=ax[0]);
ax[0].set_title('Distribution of Arrival Delay Time (Min)', fontsize=14);
ax[0].set_xlim([min(df['arr_delay'].values), max(df['arr_delay'].values)]);

sns.boxplot(df['arr_delay'], ax=ax[1]);
ax[1].set_title('Arrival Delay Time (Min)');
plt.show();
```

#### Task 6:
##### How many states cover 50% of US air traffic?

```python
def get_state(df, col, new_col):
    value=[]
    for i in df[col]:
        i = str(i)
        value.append(i[-2:])
    df[new_col] = value        
```

```python
get_state(df, 'origin_city_name', 'state_origin')
get_state(df, 'dest_city_name', 'state_dest')
origin = df[['mkt_carrier_fl_num','state_origin']].groupby('state_origin').agg('count').sort_values('mkt_carrier_fl_num',ascending=False).rename(columns={'mkt_carrier_fl_num':'nflights'})
origin['pflights'] = origin['nflights']/sum(origin['nflights'])*100
origin.reset_index(inplace=True)
top50 = []
perc = 0
for index, row in origin.iterrows():
    perc += row['pflights']
    if perc > 50:
        break
    top50.append(row['state_origin'])
```

```python
print('Number of US States comprising 50% US air traffic: ', len(top50))
print('US States: ', top50)
```

## Feature Engineering

```python
df['fl_date'] = pd.to_datetime(df['fl_date'], errors='coerce')
df['month'] = df['fl_date'].dt.month
df['day_of_week'] = df['fl_date'].dt.dayofweek
df['day_of_month'] = df['fl_date'].dt.day
df['year'] = df['fl_date'].dt.year
```

```python
df['crs_dep_time'] = df['crs_dep_time'].apply(lambda x: '{0:0>4}'.format(x))
df['crs_arr_time'] = df['crs_arr_time'].apply(lambda x: '{0:0>4}'.format(x))
df['dep_hour']=df['crs_dep_time'].str[:2]
df['arr_hour']=df['crs_arr_time'].str[:2]
```

```python
df['seasons']='spring'
df.loc[df['month'].between(6, 8), 'seasons']= 'summer'
df.loc[df['month'].between(9, 11), 'seasons']= 'autum'
df.loc[df['month'].between(12, 2), 'seasons']= 'winter'
```

```python
df.head()
```

```python
df['dep_hour']=df['dep_hour'].astype(int)
df['dep_time_during_day']='night'
df.loc[df['dep_hour'].between(12,18), 'dep_time_during_day']= 'afternoon'
df.loc[df['dep_hour'].between(19,23), 'dep_time_during_day']= 'evening'
df.loc[df['dep_hour'].between(4,11), 'dep_time_during_day']= 'morning'

df['arr_hour']=df['arr_hour'].astype(int)
df['arr_time_during_day']='night'
df.loc[df['arr_hour'].between(12,18), 'arr_time_during_day']= 'afternoon'
df.loc[df['arr_hour'].between(19,23), 'arr_time_during_day']= 'evening'
df.loc[df['arr_hour'].between(4,11), 'arr_time_during_day']= 'morning'
```

```python
df= df.drop('fl_date', axis = 1)
```

```python
df['dep_time_during_day'].value_counts()
```

```python
def flight_duration(x):
    if x <=180:
        return 'Short'
    elif x >180 and x<360:
        return 'Medium'
    elif x>360:
        return 'Long'

df['flight_duration_type']=df['air_time'].apply(lambda x: flight_duration(x))
df['flight_duration_type'].value_counts()
```

```python
df['origin_state']=df['origin_city_name'].apply(lambda x: x.split(', ')[1])# Separate State from city
df['destination_state']=df['dest_city_name'].apply(lambda x: x.split(', ')[1])
```

```python
#Find mean of depart and arrival delays based on # carriers
df['dep_delay_mean_by_carrier']=df.groupby(['op_unique_carrier'])['dep_delay'].transform(np.mean)
df['arr_delay_mean_by_carrier']=df.groupby(['op_unique_carrier'])['arr_delay'].transform(np.mean)
```

```python
df['arr_delay_cat'] = df['arr_delay'].apply(lambda x: 1 if x > 0 else 0) #ENCODE ARRIV DELAY
```

```python
# ENCODE AIRPORTS AND TAIL NUM
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
df['mkt_carrier'] = encoder.fit_transform(df[['mkt_carrier']])
df['mkt_unique_carrier'] = encoder.fit_transform(df[['mkt_unique_carrier']])
df['mkt_carrier_fl_num'] = encoder.fit_transform(df[['mkt_carrier_fl_num']])
df['op_unique_carrier'] = encoder.fit_transform(df[['op_unique_carrier']])
df['tail_num'] = encoder.fit_transform(df[['tail_num']])
df['op_carrier_fl_num'] = encoder.fit_transform(df[['op_carrier_fl_num']])
df['origin_airport_id'] = encoder.fit_transform(df[['origin_airport_id']])
df['dest_airport_id'] = encoder.fit_transform(df[['dest_airport_id']])
```

```python
#Find air traffics based on the airports
df['origin_airport_taffic']=df.groupby(['origin_airport_id'])['flights'].transform(np.sum)
df['dest_airport_traffic']=df.groupby(['dest_airport_id'])['flights'].transform(np.sum)
```

```python
dummy_durtypes=pd.get_dummies(df[['flight_duration_type', 'dep_time_during_day', 'arr_time_during_day', 'seasons']])
df=pd.concat([df, dummy_durtypes], axis=1).drop(['flight_duration_type', 'dep_time_during_day', 'arr_time_during_day', 'seasons'], axis=1)
```
